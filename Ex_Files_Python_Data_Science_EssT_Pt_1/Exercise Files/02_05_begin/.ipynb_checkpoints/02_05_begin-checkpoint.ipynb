### Grouping data by column index

```python
#subgrouping
#useful for finding common subgroup (sort of like a common denominator) within the data

address = 'C:/Users/guthrie/Documents/GitHub/PythonEssentialsForDataSciencePart1/Ex_Files_Python_Data_Science_EssT_Pt_1/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
#print first 5 rows of dataframe
cars.head()
```

```python
#group (i.e. pivot) based on cylinder and then averages each category 
#for the vehicles with that number of cylinders
cars_groups = cars.groupby(cars['cyl'])
cars_groups.mean().round(2)
```
